# Min Max Scaler
    ~ Min=0, Max=1
    ~ (0 to 1 range value)=(original value-min. value)/(max.value-min. value)
    ~ Scaling is mainly used for regressions because we get different weights in the data.

```python
import pandas as pd
```

```python
df=pd.read_table('CommViolPredUnnormalizedData.txt',sep=',',na_values='?')
df
```

```python
req_columns=[5,6]+list(range(11,26))+list(range(32,103))+[145]
```

```python
df.shape
```

```python
df=df.iloc[:,req_columns]
```

```python
df.shape
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
X=df[df.columns[:-1]]
X.shape
```

```python
y=df[df.columns[-1]]
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

### fit and transform the training feature data for scaler object

```python
X_train_scaled=scaler.fit_transform(X_train)
```

### transform the testing features using training feature scaler object
    test features hvae to be scaled using traing scaler object only i.e, scaler=MinMaxscaler()

```python
X_test_scaled=scaler.transform(X_test)
```

## Train and test linear model without scaled data and with scaled data

# without scaled data

```python
from sklearn.linear_model import LinearRegression
```

```python
linmodel=LinearRegression()
```

```python
linmodel.fit(X_train,y_train)   ## training score
linmodel.score(X_train,y_train)
```

```python
linmodel.score(X_test,y_test)   ## testing score
```

# with scaled data

```python
linmodel1=LinearRegression()
linmodel.fit(X_train_scaled,y_train)
```

```python
linmodel.score(X_train_scaled,y_train) ## Training score
```

```python
linmodel.score(X_test_scaled,y_test)  ## Testing score
```

#  Ridge model

## Without scaling

```python
from sklearn.linear_model import Ridge
```

```python
ridgemodel=Ridge(alpha=1)
ridgemodel.fit(X_train,y_train)
```

```python
ridgemodel.score(X_train,y_train)  ## Training score
```

```python
ridgemodel.score(X_test,y_test)  ## testing score
```

# With scaling

```python
ridgemodel1=Ridge(alpha=1)
ridgemodel1.fit(X_train_scaled,y_train)
ridgemodel1.score(X_train_scaled,y_train)   ## Training score
```

```python
ridgemodel1.score(X_test_scaled,y_test)  ## testing score
```

#  POLYNOMIAL FEATURES
    When we increase the degree, the features also will increase.
    Polynomial features are used for modifications that are needed.

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
from sklearn.datasets import make_friedman1
```

```python
X_f1,y_f1=make_friedman1(n_samples=100,n_features=7,random_state=42)
```

```python
# create object for polynomial features
poly=PolynomialFeatures()
```

```python
X_f1_poly=poly.fit_transform(X_f1)
```

```python
X_f1.shape ## after applying polynomial features this is the final shape
```

```python
X_f1_poly.shape ## before applying polynomial features this is the final shape
```

```python
from sklearn.model_selection import train_test_split
```

### Without poly

```python
X_train,X_test,y_train,y_test=train_test_split(X_f1,y_f1,test_size=0.30,random_state=42)
```

### With poly

```python
X_train_poly,X_test_poly,y_train,y_test=train_test_split(X_f1_poly,y_f1,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linmodel=LinearRegression()
```

# without poly

```python
linmodel.fit(X_train,y_train)
print('Training score is ')
print(linmodel.score(X_train,y_train))
print('Testing score is ')
print(linmodel.score(X_test,y_test))
```

# with poly

```python
linmodelpoly=LinearRegression()
linmodel.fit(X_train_poly,y_train)
print('Training score is ')
print(linmodel.score(X_train_poly,y_train))
print('Testing score is ')
print(linmodel.score(X_test_poly,y_test))
```

# Verifying Ridge model with polynomial features

## without poly

```python
ridgemodel=Ridge(alpha=1)
ridgemodel.fit(X_train,y_train)
```

```python
ridgemodel.score(X_train,y_train)
```

```python
ridgemodel.score(X_test,y_test)
```

# With poly

```python
ridgemodelpoly=Ridge(alpha=1)
ridgemodelpoly.fit(X_train_poly,y_train)
print('Training score is ')
print(ridgemodelpoly.score(X_train_poly,y_train))
print('Testing score is ')
print(ridgemodelpoly.score(X_test_poly,y_test))
```
