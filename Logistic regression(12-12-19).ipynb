# Logistic Regression:
    ~ It is a classification model.
    ~ Application- used for classification(0 or 1)
    ~ Like L2 regularization in ridge regression, here C is used for regularization(control)
    ~ We can use for multiple classifications also.
    

```python
import pandas as pd
```

```python
df=pd.read_table('fruit_data_with_colors.txt')
df
```

## applying binary classification(apples and not apples)

```python
df.loc[df['fruit_label']!=1,'fruit_label']=0
df.head()
```

```python
df['fruit_label'].value_counts()
```

```python
X=df[['mass','width','height']]
X
```

```python
y=df['fruit_label']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
X_test.head()
```

```python
y.head()
```

```python
logmodel.predict([[150,7.1,7.9]])
```

```python
logmodel.score(X_train,y_train)
```

```python
logmodel.score(X_test,y_test)
```

# breast cancer dataset

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
df=pd.DataFrame(cancer['data'])
df.columns=cancer['feature_names']
df.head()
```

```python
df['Target']=cancer['target']
df.head()
```

```python
df.isnull().sum()
```

```python
X=df[['mean radius','mean texture','mean perimeter','mean area','mean smoothness','mean compactness','mean concavity','mean concave points','mean symmetry']]
X
```

```python
y=df['Target']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
logmodel3=LogisticRegression()
logmodel3.fit(X_train,y_train)
```

```python
logmodel3.score(X_train,y_train)
```

```python
logmodel3.score(X_test,y_test)
```

```python
logmodel3.predict([[19.69,21.25,130.00,1203.0,0.10960,0.15990,0.1974,0.12790,0.2069]])
```
