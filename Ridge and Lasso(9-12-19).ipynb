```python
import pandas as pd
```

```python
df=pd.read_table('CommViolPredUnnormalizedData.txt',sep=',',na_values='?')  ## na_values=? represents '?' data is filled with Nan
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
X=df[['population','householdsize','racepctblack','racePctWhite','racePctAsian','racePctHisp','agePct12t21','agePct12t29','agePct16t24','medIncome','pctWWage']]

X
```

```python
y=df['ViolentCrimesPerPop']

y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridgemodel=Ridge(alpha=1)
ridgemodel.fit(X_train,y_train)
```

```python
ridgemodel.score(X_test,y_test)
```

```python
ridgemodel.coef_
```

```python
alpha=[1,10,20,50,100]
scores={}
for a in alpha:
    ridgemodel=Ridge(alpha=a)
    ridgemodel.fit(X_train,y_train)
    scores[a]=ridgemodel.score(X_test,y_test)
    
```

```python
scores
```

# Lasso regression
    ~ In lasso reg L1 regularization is used.
    ~ Sum of absolute coefficients
    ~ Only few variables with medium/large effect- use Lasso
    ~ This reg can also be used to find out best features(Ex:- given below)

```python
from sklearn.linear_model import Lasso
```

```python
lassomodel=Lasso(alpha=1)
```

```python
lassomodel.fit(X_train,y_train)
```

```python
lassomodel.score(X_test,y_test)
```

```python
lassomodel.coef_
```

```python
len(X.columns)
```

```python
## We can find best features with non 0 and 0 , 0 elements mean that feature is not effected
sum(lassomodel.coef_!=0)  ## if there are 0's in the array means that some features are not required
```
